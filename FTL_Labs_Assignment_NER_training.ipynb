```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

# Code to parse ALL pages of ALL pdfs and output text

```python
prob_resumes = []
for name in glob.glob(r'./censored_ubuntu_fix/*.pdf'):
    #print(name)
    all_text = ""
    with pdfplumber.open(name) as pdf:
        for pdf_page in pdf.pages:
            single_page_text = pdf_page.extract_text()
            try:
                all_text = all_text + '\n' + single_page_text
            except (AttributeError,TypeError):
                prob_resumes.append(name)
                break
        if all_text.count("cid:") > 2:
            prob_resumes.append(name)
        else:
            if name in prob_resumes:
                pass
            else:
                print(color.BOLD + name + color.END)
                print(all_text)
                print("\n")
                print("___"*40)

```

```python
print("The resumes with problems being parsed are: ")
for x in prob_resumes:
    print(x)
```

# Using Spacy NER

```python
resume_dict = {}
for name in glob.glob(r'./censored_ubuntu_fix/*.pdf'):
    if name not in prob_resumes:
        #print(name)
        all_text = ""
        with pdfplumber.open(name) as pdf:
            for pdf_page in pdf.pages:
                single_page_text = pdf_page.extract_text()
                all_text = all_text + '\n' + single_page_text
        resume_dict[name] = all_text
```
