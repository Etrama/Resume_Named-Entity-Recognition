# Check whether all files can be parsed

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

# Code to parse ALL pages of a single pdf and output text

```python
name = r"C:\Users\kmoudgalya\Desktop\FTL\ML Assignment test dataset\censored\resume_10.pdf"
all_text = ""
with pdfplumber.open(name) as pdf:
    for pdf_page in pdf.pages:
        single_page_text = pdf_page.extract_text()
        try:
            all_text = all_text + '\n' + single_page_text
        except (AttributeError,TypeError):
            prob_resumes.append(name)
            break
    if all_text.count("cid:") > 2:
        prob_resumes.append(name)
    else:
        if name in prob_resumes:
            pass
        else:
            print(color.BOLD + name + color.END)
            print(all_text)
            print("\n")
            print("___"*40)
```

```python
import PyPDF2
import requests
import json

filename = r"C:\Users\kmoudgalya\Desktop\FTL\ML Assignment test dataset\censored\resume_1.pdf"
#file = open('Resume.pdf', 'rb')
pdf_reader = PyPDF2.PdfFileReader(filename)
pageObj = pdf_reader.getPage(0)
print("Total Pages : {} ".format(pdf_reader.numPages))

resume = pageObj.extractText()
#url = "https://api.iki.ai/api/skills_extraction/"
#url =  "https://dragon.pdf2go.com/api/user/"
url =  "https://dragon.pdf2go.com/api/user/me?_=1603980530355"



payload = {
    "text":str(resume[0:2000])
}


headers = {
    'Content-Type': 'application/json'
}

print(len(resume))


r = requests.get(url=url, headers=headers, data=json.dumps(payload))
print(r.json())
```
