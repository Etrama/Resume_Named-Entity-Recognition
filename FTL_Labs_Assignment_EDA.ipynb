# Check whether all files can be parsed

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
prob_resumes = []
for name in glob.glob(r'C:\Users\kmoudgalya\Desktop\FTL\ML Assignment test dataset\censored\*.pdf'): 
    with pdfplumber.open(name) as pdf:
        page = pdf.pages[0]
        text = page.extract_text()
        print(color.BOLD + name + color.END)
        print(text)
        print("\n")
        print("___"*40)
```

```python
prob_resumes = []
for name in glob.glob(r'C:\Users\kmoudgalya\Desktop\FTL\ML Assignment test dataset\censored\*.pdf'): 
    with pdfplumber.open(name) as pdf:
        page = pdf.pages[0]
        text = page.extract_text()
        try:
            if text.count("cid:") > 2:
                prob_resumes.append(name)
                continue
            else:
                print(color.BOLD + name + color.END)
                print(text)
                print("\n")
                print("___"*40)
        except AttributeError:
            prob_resumes.append(name)
            continue
```

```python
print("The resumes with problems being parsed are: ")
for x in prob_resumes:
    print(x)
```

```python
import os
import tempfile
from pdf2image import convert_from_path
 
filename = r"C:\Users\kmoudgalya\Desktop\FTL\ML Assignment test dataset\censored\resume_1.pdf"
 
with tempfile.TemporaryDirectory() as path:
     images_from_path = convert_from_path(filename, output_folder=path, last_page=1, first_page =0)
 
base_filename  =  os.path.splitext(os.path.basename(filename))[0] + '.jpg'     
 
save_dir = 'your_saved_dir'
 
for page in images_from_path:
    page.save(os.path.join(save_dir, base_filename), 'JPEG')
```
